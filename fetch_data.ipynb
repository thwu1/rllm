```python
from episodic import ContextStore, ContextFilter
from rllm.sdk import LLMTracer
```

```python
store = ContextStore(
    endpoint="http://localhost:8000",
    api_key="your-api-key-here",
)

tracer = LLMTracer(store, project="rllm-agent-omni-engine")

result = await tracer.query_traces(
    # since="1h",
    limit=512
)
```

```python
import json

with open("result.json", "w") as f:
    for trace in result:
        f.write(json.dumps(trace.data) + "\n")
```
